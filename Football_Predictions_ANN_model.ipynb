### Importing the libraries

## Part 1 - Data Preprocessing

### Importing the dataset

The neccesery libraries for ANN is numpy, pandas for data management and handling, tensorflow for the ANN nodes creation and scikit learn for neccesery machine learn functions

```python
import numpy as np
import tensorflow as tf
import pandas as pd
```

```python
tf.__version__
```

The main dataset is the seria_data_bs.csv, it contains all the independent variables and the dependent football game outcome. The dataset is then split to the x independent variables and the y dependent variable set (H for home win, D for Draw and A for away win). The iloc function is used and it takes all the rows [:] and all the columns except the last one [:-1] for the x set and again all the rows [:] and just the last column [-1] for the y set

```python
dataset = pd.read_csv('seria_data_bs.csv')
x = dataset.iloc[:, :-1].values
y = dataset.iloc[:, -1].values
```

```python
print(x)
```

```python
print(y)
```

Encoding categorical dependent variable data

One - Hot encoding our dependent variable. Our dependent variable has 3 outcomes: Home win (H), Draw (D) and away win (A). We need to performe one-hot encoding on these categorical data in order to use them in the model as the dependent variable. After this encoding H win is encoded [0.0.1], Draw as [0.1.0] and Away win [1.0.0]. 

```python
from sklearn.compose import ColumnTransformer
from sklearn.preprocessing import OneHotEncoder

# assuming your original dataframe is called 'y' and the column to be one-hot encoded is called 'outcome'
ct = ColumnTransformer(transformers=[('encoder', OneHotEncoder(), [0])], remainder='drop')

# reshape 'y' to a 2D numpy array with a single column
y_reshaped = y.reshape(-1, 1)

# transform the 'outcome' column to a numpy array and perform one-hot encoding using the ColumnTransformer
y_encoded = np.array(ct.fit_transform(y_reshaped))
```

```python
print(y_encoded)
```

### Splitting the dataset into the Training set and Test set

We will use the trainsplit function from the model selection module of sklearn. Test size is set at 0.2 meaning a 80% - 20% split. Random_state is set to 1. 

```python
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y_encoded, test_size = 0.2, random_state = 1)
```

```python
print(x_train)
```

```python
print(x_test)
```

```python
print(y_train)
```

```python
print(y_test)
```

### Feature Scaling

```python
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x_train = sc.fit_transform(x_train)
x_test = sc.transform(x_test)
```

```python
print(x_train)
```

```python
print(x_test)
```

## Part 2 - Building the ANN

### Initializing the ANN

```python
ann = tf.keras.models.Sequential()
```

### Adding the input layer and the first hidden layer

```python
ann.add(tf.keras.layers.Dense(units=6, activation = "relu"))
```

### Adding the second hidden layer

```python
ann.add(tf.keras.layers.Dense(units=6, activation = "relu"))
```

### Adding the output layer

Here we are adding the output layer and since our dependent variable is the outcome of the game represented as three independent classes (H, D and A) that are one hot encoded we need the "units" parameter to be set to three because we need three neurons to produce a binary outcome that encodes each class. The "activation" parameter of the output layer also needs to be set to a sigmoid activation function. That is because a sigmoid function produces the predictions as well as the probabilities of each outcome.

```python
ann.add(tf.keras.layers.Dense(units=3, activation = "softmax"))
```

## Part 3 - Training the ANN

### Compiling the ANN

```python
ann.compile(optimizer= "adam", loss= "categorical_crossentropy", metrics= ["accuracy"])
```

### Training the ANN on the Training set

```python
ann.fit(x_train, y_train, batch_size = 32, epochs = 10000)
```

## Part 4 - Making the predictions and evaluating the model

### Predicting the result of a single observation

**Solution**

```python
ann.predict([[]])
```

Therefore, our ANN model predicts that this customer stays in the bank!

**Important note 1:** Notice that the values of the features were all input in a double pair of square brackets. That's because the "predict" method always expects a 2D array as the format of its inputs. And putting our values into a double pair of square brackets makes the input exactly a 2D array.

**Important note 2:** Notice also that the "France" country was not input as a string in the last column but as "1, 0, 0" in the first three columns. That's because of course the predict method expects the one-hot-encoded values of the state, and as we see in the first row of the matrix of features X, "France" was encoded as "1, 0, 0". And be careful to include these values in the first three columns, because the dummy variables are always created in the first columns.

### Predicting the Test set results

```python
y_pred = ann.predict(x_test)
y_pred = (y_pred > 0.5)
y_pred = y_pred.astype(int)
print(y_pred)
```

```python
# Define the mapping of class indices to labels
label_map = {0: "A", 1: "D", 2: "H"}

# Transform the matrix y_pred to a matrix of labels
y_pred_labels = np.argmax(y_pred, axis=1)
y_pred_labels = [label_map[i] for i in y_pred_labels]

print(y_pred_labels)
```
